```python
import pandas as pd

from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

## Load Data (speed 250) from json files

```python
lane1_position_updates = pd.read_json(path_or_buf='lane1_speed250_position_updates.json', orient='records')
lane2_position_updates = pandas.read_json('lane2_speed250_position_updates.json', orient='records')
lane3_position_updates = pandas.read_json('lane3_speed250_position_updates.json', orient='records')
lane1_transition_updates = pandas.read_json('lane1_speed250_transition_updates.json', orient='records')
lane2_transition_updates = pandas.read_json('lane2_speed250_transition_updates.json', orient='records')
lane3_transition_updates = pandas.read_json('lane3_speed250_transition_updates.json', orient='records')
lane4_transition_updates = pandas.read_json('lane4_speed250_transition_updates.json', orient='records')
```

```python
display_side_by_side(lane1_position_updates[["car_no", "offset", "real_piece_id", "piece_location"]], 
                     lane2_position_updates[["car_no", "offset", "real_piece_id", "piece_location"]],
                     lane3_position_updates[["car_no", "offset", "real_piece_id", "piece_location"]])
```
